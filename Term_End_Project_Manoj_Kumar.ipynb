```python
# Applying text cleaning techniques for removing'\[.*?\]', '[%s]' '\w*\d\w*'
import re
import string

def clean_text_round1(comment_clean):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    comment_clean = comment_clean.lower()
    comment_clean = re.sub('\[.*?\]', '', comment_clean)
    comment_clean = re.sub('[%s]' % re.escape(string.punctuation), '', comment_clean)
    comment_clean = re.sub('\w*\d\w*', '', comment_clean)
    return comment_clean

round1 = lambda x: clean_text_round1(x)
df['comment_clean'] = df.comment_clean.apply(round1)
df.comment_clean
```

```python
# Applying text cleaning techniques for removing '[‚Äò‚Äô‚Äú‚Äù‚Ä¶]' , '\n'
def clean_text_round2(comment_clean):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    comment_clean = re.sub('[‚Äò‚Äô‚Äú‚Äù‚Ä¶.]', '', comment_clean)
    comment_clean = re.sub('\n', '', comment_clean)
    return comment_clean

round2 = lambda x: clean_text_round2(x)
df['comment_clean'] = df.comment_clean.apply(round2)
df.comment_clean
```

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(comment_preprocessed):
    new_message = []
    for w in comment_preprocessed.split():
        if w.upper() in chat_words_list:
            new_message.append(chat_words_map_dict[w.upper()])
        else:
            new_message.append(w)
    return " ".join(new_message)

chat_words_conversion("This is a test sentence to check if chat words writen in short form can be written clearly with full form words such as: one minute BRB")
```

```python
import nltk
#lexicon 1 for sentiment analysis
from nltk.corpus import opinion_lexicon 

##dictionary of positive and negative words from opinion lexicon
pos_list=set(opinion_lexicon.positive())
neg_list=set(opinion_lexicon.negative())

##tokensier to be used
from nltk.tokenize import treebank  

# lexicon 2 for sentiment analysis 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#construct a tokenizer 

tokenizer = treebank.TreebankWordTokenizer()

# takes a string as input, tokenizes it, 
#counts the number of positive and negative words in the string, 
#and calculates their difference as sentiment

def sentiment(sentence):
    senti=0
    words = [word for word in tokenizer.tokenize(sentence)]
    for word in words:
        if word in pos_list:
            senti += 1
        elif word in neg_list:
            senti -= 1
    return senti

# apply it to every document 
df['sentiment_opionion_lexicon']=df['comment'].apply(sentiment)
```
